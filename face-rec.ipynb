```python
import matplotlib.pyplot as plt
```

```python
def show_images(images) -> None:
    n: int = len(images)
    f = plt.figure()
    for i in range(n):
        f.add_subplot(1, n, i + 1)
        plt.imshow(images[i])
    plt.show(block=True)
```

```python
import glob
import cv2
```

```python
images = [cv2.imread(file) for file in glob.glob("imgdata/*")]
```

```python
images
```

```python
show_images(images)
```
